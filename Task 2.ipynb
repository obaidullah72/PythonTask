```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = r'F:\TASK PYthon\Annual_Living_Cost.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand data structure
print("Original Data Preview:")
print(data.head())

# 1. Drop duplicate rows
data = data.drop_duplicates()
print(f"Data shape after dropping duplicates: {data.shape}")

# 2. Handle missing values based on data type
#   a. For numeric columns, fill missing values with the median.
#   b. For categorical columns, fill missing values with the mode.

numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Fill numeric columns with median
data[numeric_cols] = data[numeric_cols].apply(lambda col: col.fillna(col.median()))

# Fill categorical columns with mode
for col in categorical_cols:
    mode_value = data[col].mode()[0]
    data[col] = data[col].fillna(mode_value)

# 4. Drop remaining rows with NaN values after all previous handling
data_cleaned = data.dropna()

# Display the cleaned data preview and final shape
print("Cleaned Data Preview:")
print(data_cleaned.head())
print(f"Final data shape after cleaning: {data_cleaned.shape}")

# 5. Save the cleaned data to a new CSV file
output_file_path = r'F:\TASK PYthon\Annual_Living_Cost_Cleaned.csv'
data_cleaned.to_csv(output_file_path, index=False)

print("Cleaned data has been saved to:", output_file_path)
```

```python
import pandas as pd
import numpy as np

def clean_data(file_path, output_file_path):
    # Load the CSV file
    data = pd.read_csv(file_path)

    # Display the first few rows to understand data structure
    print(f"Original Data Preview for {file_path}:")
    print(data.head())

    # 1. Drop duplicate rows
    data = data.drop_duplicates()
    print(f"Data shape after dropping duplicates: {data.shape}")

    # 2. Handle missing values based on data type
    #   a. For numeric columns, fill missing values with the median.
    #   b. For categorical columns, fill missing values with the mode.

    numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = data.select_dtypes(include=['object']).columns

    # Fill numeric columns with median
    data[numeric_cols] = data[numeric_cols].apply(lambda col: col.fillna(col.median()))

    # Fill categorical columns with mode
    for col in categorical_cols:
        mode_value = data[col].mode()[0]
        data[col] = data[col].fillna(mode_value)

    # 3. Drop remaining rows with NaN values after all previous handling
    data_cleaned = data.dropna()

    # Display the cleaned data preview and final shape
    print("Cleaned Data Preview:")
    print(data_cleaned.head())
    print(f"Final data shape after cleaning: {data_cleaned.shape}")

    # 5. Save the cleaned data to a new CSV file
    data_cleaned.to_csv(output_file_path, index=False)
    print("Cleaned data has been saved to:", output_file_path)

# List of file paths to clean
files_to_clean = [
    (r'F:\TASK PYthon\Annual_Living_Cost.csv', r'F:\TASK PYthon\Annual_Living_Cost_Cleaned.csv'),
    (r'F:\TASK PYthon\Employed_Income.csv', r'F:\TASK PYthon\Employed_Income_Cleaned.csv'),
    (r'F:\TASK PYthon\Employed_People.csv', r'F:\TASK PYthon\Employed_People_Cleaned.csv'),
    (r'F:\TASK PYthon\Homeless.csv', r'F:\TASK PYthon\Homeless_Cleaned.csv'),
    (r'F:\TASK PYthon\Housing_Affordability.csv', r'F:\TASK PYthon\Housing_Affordability_Cleaned.csv'),
    (r'F:\TASK PYthon\Housing_Cost.csv', r'F:\TASK PYthon\Housing_Cost_Cleaned.csv')
]

# Process each file
for input_file, output_file in files_to_clean:
    clean_data(input_file, output_file)
```
